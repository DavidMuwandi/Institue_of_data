```python

#Setting the Image dimension and source folder for loading the dataset
IMG_WIDTH=150
IMG_HEIGHT=150
img_folder= r'.\dataset\waste_classification'

```

```python
# create function to create the image data and the labels from the images in the folder

def create_dataset(img_folder):
   
    img_data_array=[]
    class_name=[]
   
    for dir1 in os.listdir(img_folder):
        for file in os.listdir(os.path.join(img_folder, dir1)):
       
            image_path= os.path.join(img_folder, dir1,  file)  
            image= cv2.imread( image_path, cv2.COLOR_BGR2RGB)                                   #Read the image file from the folder and convert it to the right color format.
            image=cv2.resize(image, (IMG_HEIGHT, IMG_WIDTH),interpolation = cv2.INTER_AREA)     #Resize the image based on the input dimension required for the model
            image=np.array(image)
            image = image.astype('float32')                                                      #Convert the image to a Numpy array with float32 as the datatype
            image /= 255                                                                         # Normalize the image array values for faster convergence
            img_data_array.append(image)
            class_name.append(dir1)
    return img_data_array, class_name
# extract the image array and class name
img_data, class_name =create_dataset('.\dataset\waste_classification')
```

```python
#Converting text labels to numeric codes
#Create a dictionary for all unique values for the classes
target_dict={k: v for v, k in enumerate(np.unique(class_name))}
target_dict
```

```python
#Convert the class_names to their respective numeric value based on the dictionary
target_val=  [target_dict[class_name[i]] for i in range(len(class_name))]
y = target_val
```

```python
# check data type for y
type(y)
```

```python
# Numpy array as input
#X = img_data
X = np.array(img_data, np.float32)
```

```python
# check shape for X
X.shape
```

```python
# check dististribution of classes
plt.figure(figsize = (10,8))
sns.set_context('talk')
plt.title('Class Bar Graph')
sns.despine(right = True, top=True)
sns.countplot(y)
plt.xticks(ticks = [0,1,2,3,4,5], labels =  ['cardboard','glass','metal','paper','plastic','trash'], rotation = 30)
plt.savefig('Class_bar_graph.png')
plt.show()
```

```python
# calculate class weight -----> inverse of number of samples (INS)
plt.figure(figsize = (10,8))
sns.set_context('talk')
hist, _ = np.histogram(y, bins=np.arange(7)-0.5)
loss_weight = hist.min()/hist
loss_weight = {i:loss_weight[i] for i in range(6)}

plt.bar(np.arange(6), [loss_weight[i] for i in range(6)])
plt.xlabel('Class')
plt.ylabel('class_weight')
plt.title('class_weight bar graph')
sns.despine(top = True, right = True)
plt.xticks(ticks = [0,1,2,3,4,5], labels =  ['cardboard','glass','metal','paper','plastic','trash'], rotation = 30)
plt.savefig('class_weight_graph.png')
plt.show()
```

```python
print(loss_weight)
```

```python
# Alternative method for class weights

from sklearn.utils import class_weight

#np.array(y)

class_weights = class_weight.compute_class_weight('balanced',
                                                 np.unique(np.array(y)),
                                                 y)
# Normalised
class_weights = class_weights/class_weights.max()
class_weights

#model.fit(X_train, y_train, class_weight=class_weights)
```

```python
# one hot encode labels

y = pd.get_dummies(y)
y
```

```python
y.value_counts()
```

```python
# convert y back into array
y = np.array(y,np.float32)
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train ,y_test = train_test_split(X,y, train_size = 0.8, test_size = 0.2, random_state = 45, stratify = y)
```

```python
# Create genarator for image augmentatation


datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    vertical_flip = True,
    validation_split = 0.2,
    #zoom_range=0.1,
    #brightness_range=[0.8,1.2]

)


train_generator = datagen.flow(X_train, y_train, seed = 45, subset = 'training')

validation_generator = datagen.flow(X_train, y_train, seed = 45, subset = 'validation')
```

# CNN Model

```python
# Create model architecture

num_classes = len(target_dict)

model2 = Sequential()
model2.add(Conv2D(48, kernel_size = 3, activation = 'relu', padding = 'same', input_shape = (IMG_WIDTH, IMG_HEIGHT, 3)))
model2.add(Conv2D(48, kernel_size = 3, activation = 'relu'))
model2.add(MaxPool2D(pool_size = (2, 2)))
model2.add(Dropout(0.25))
model2.add(Conv2D(96, kernel_size = 3, activation = 'relu', padding = 'same'))
model2.add(Conv2D(96, kernel_size = 3, activation = 'relu'))
model2.add(MaxPool2D(pool_size = (2, 2)))
model2.add(Dropout(0.25))
model2.add(Conv2D(192, kernel_size = 3, activation = 'relu', padding = 'same'))
model2.add(Conv2D(192, kernel_size = 3, activation = 'relu'))
model2.add(MaxPool2D(pool_size = (2, 2)))
model2.add(Dropout(0.25))
model2.add(Flatten())
model2.add(Dense(512, activation = 'relu'))
model2.add(Dropout(0.5))
model2.add(Dense(256, activation = 'relu'))
model2.add(Dropout(0.5))
model2.add(Dense(num_classes, activation = 'softmax'))
```

```python
# Compile model

model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'], loss_weights = class_weights)

model2.summary()
```

```python

# early stopping with patience, instantiate callbacklist

es = EarlyStopping(monitor='val_loss', verbose=1, patience = 10,restore_best_weights=True)

model2_filepath = 'cnn_model2.hdf5'
model2_checkpoint = tf.keras.callbacks.ModelCheckpoint(model2_filepath, monitor='val_accuracy', verbose=1, save_best_only=True)
callbacklist = [model2_checkpoint, es]

#fit model

number_of_epochs = 120
history = model2.fit(train_generator, epochs=number_of_epochs, validation_data = validation_generator, callbacks= callbacklist,verbose = 1, batch_size = 20)

# load saved weights instead of refitting model
#model2.load_weights("./my_cnn_model2.hdf5")
```

# VGG16

```python
# loading the vgg16 model
from keras.applications.vgg16 import VGG16

```

```python
# model architecture

vgg_epoch = 120

vgg16_model = VGG16(pooling='avg', weights='imagenet', include_top=False, input_shape=(IMG_WIDTH, IMG_HEIGHT, 3))
for layers in vgg16_model.layers:
            layers.trainable=False
last_output = vgg16_model.layers[-1].output
vgg_x = Flatten()(last_output)
vgg_x = Dense(128, activation = 'relu')(vgg_x)
vgg_x = Dropout(0.5)(vgg_x)
vgg_x = Dense(6, activation = 'softmax')(vgg_x)
vgg16_final_model = Model(vgg16_model.input, vgg_x)
vgg16_final_model.compile(loss = 'categorical_crossentropy', optimizer= 'adam', metrics=['acc'], loss_weights = class_weights)

```

```python
vgg16_final_model.summary()
```

```python
# Instatiate callbacklist
number_of_epochs = vgg_epoch
vgg16_filepath = 'vgg_16.hdf5'
vgg_checkpoint = tf.keras.callbacks.ModelCheckpoint(vgg16_filepath, monitor='val_acc', verbose=1, save_best_only=True)

vgg_early_stopping = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10,restore_best_weights=True)
callbacklist = [vgg_checkpoint, vgg_early_stopping]

# fit model
vgg16_history = vgg16_final_model.fit(train_generator, epochs = number_of_epochs ,validation_data = validation_generator,callbacks=callbacklist,verbose=1, batch_size = 20)

#load saved weights
#vgg16_final_model.load_weights("./my_vgg16.hdf5")
```

# Resnet50

```python
# example of loading the resnet50 model
from keras.applications.resnet50 import ResNet50

# load model
new_input = IMG_WIDTH, IMG_HEIGHT, 3
ResNet50_model = ResNet50(weights='imagenet', include_top=False, input_shape=new_input, classes=6)
```

```python

for layers in ResNet50_model.layers:
    layers.trainable=False


resnet50_x = Flatten()(ResNet50_model.output)
resnet50_x = Dense(256,activation='relu')(resnet50_x)
resnet50_x = Dense(6,activation='softmax')(resnet50_x)
resnet50_x_final_model = Model(inputs=ResNet50_model.input, outputs=resnet50_x)
resnet50_x_final_model.compile(loss = 'categorical_crossentropy', optimizer= 'adam', metrics=['acc'], loss_weights = class_weights)
```

```python
resnet50_x_final_model.summary()
```

```python
# instatiate callbacklist 

number_of_epochs = 120
resnet_filepath = 'resnet50.hdf5'
resnet_checkpoint = tf.keras.callbacks.ModelCheckpoint(resnet_filepath, monitor='val_acc', verbose=1, save_best_only = True)
resnet_early_stopping = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10,restore_best_weights=True)
callbacklist = [resnet_checkpoint,resnet_early_stopping]


# fit model
resnet50_history = resnet50_x_final_model.fit(train_generator, epochs = number_of_epochs ,validation_data = (validation_generator),callbacks=callbacklist,verbose=1, batch_size = 20)


#load saved weights
#model.load_weights("./my_resnet50.hdf5")
```
