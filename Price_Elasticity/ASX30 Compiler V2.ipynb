```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
List =['BHP.AX.csv', 'CBA.AX.csv','RIO.AX.csv','SHL.AX.csv','SUN.AX.csv','REA.AX.csv','RHC.AX.csv','IAG.AX.csv','ASX.AX.csv','STO.AX.csv','QBE.AX.csv', 'SYD.AX.csv','BXB.AX.csv','SCG.AX.csv','NCM.AX.csv','COL.AX.csv' ,'ALL.ax.csv', 'AMC.AX.csv', 'GMG.AX.csv', 'WPL.AX.csv', 'FMG.AX.csv', 'TCL.AX.csv', 'TLS.AX.csv', 'WES.AX.csv', 'MQG.AX.csv','WOW.AX.csv','ANZ.AX.csv', 'NAB.AX.csv','WBC.AX.csv', 'CSL.AX.csv']
```

```python
df_ASX = pd.DataFrame([])

for i in List:
    df = pd.read_csv(i)
    df_ASX = df_ASX.append(df)
    
```

```python
print(df_ASX)
```

```python
df_ASX.to_csv('ASX30.csv')
```

```python
plt.figure(figsize = (50,50))
sns.set_context('paper')
sns.lineplot(x = df_ASX.Date, y = df_ASX.Open, data = df_ASX, hue = 'Stock')
plt.show()
```
